## DAILY NIGERIAN COVID-19 UPDATE

**TIMELINE OF CONFIRMED COVID-19 CASES IN NIGERIA 

```python
import matplotlib.pyplot as plt
```

```python
weeks = ['week 1', 'week 2', 'week 3', 'week 4', 'week 5', 'week 6']
weekly = [1, 0, 1, 20, 44, 73]
plt.plot(weeks, weekly, color='r')
plt.show()
```

```python
plt.bar(weeks, weekly, color='c')
plt.show()
```

The Cases increase on weekly basis

At week 1, only a case was confirmed.
The following week no case was recorded.
At week 3, another case was recorded.
In the fourth week, 20 new cases were confirmed.
In the fifth week, 44 new cases.
In the current week, 73 cases have been confirmed.

NB: Weeks are run in a 5-day span.
